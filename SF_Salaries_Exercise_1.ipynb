*** What was the average (mean) BasePay of all employees per year? (2011-2014) ? ***

** What are the top 5 most common jobs? **

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

*** How many people have the word Chief in their job title? (This is pretty tricky) ***

```python
df[df['JobTitle'].str.contains('Chief')]
```

```python
df[df['JobTitle'].str.upper().str.contains('CHIEF')].shape
```

```python
len(df[df['JobTitle'].apply(lambda x: 'chief' in x.lower())])
```

** Bonus: Is there a correlation between length of the Job Title string and Salary? **

```python
# korelasi 1 = hubungan tinggi, korelasi 0 = tidak berhubungan, korelasi 01 = hubungan berkebalikan
```

```python
# Hitung panjangnya Job Title
df['title_len'] = df['JobTitle'].str.len()
```

```python
df[['title_len','TotalPayBenefits']].corr()
# -0.036 berarti mendekati 0 berarti tidak berkorelasi. 
```

```python
df[['Benefits','TotalPayBenefits']].corr()
```
